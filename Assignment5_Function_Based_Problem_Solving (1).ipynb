```python
# Problem 3: Agriculture - Rainfall Adequacy Checker
print("Problem 3: Agriculture - Rainfall Adequacy Checker")
print("-"*60)

def check_rainfall(rainfall_data, required_rainfall=100):
    """Check if rainfall is adequate for crops"""
    total_rainfall = sum(rainfall_data)
    avg_rainfall = total_rainfall / len(rainfall_data)

    if total_rainfall >= required_rainfall:
        status = "Adequate Rainfall"
    else:
        status = "Insufficient Rainfall"

    return total_rainfall, avg_rainfall, status

# Test
rainfall = [25, 30, 28, 35]
total, avg, status = check_rainfall(rainfall)
print(f"Total Rainfall: {total} mm")
print(f"Average Rainfall: {avg:.2f} mm")
print(f"Status: {status}")
print()

# Problem 4: Social Media - Duplicate Account Detection
print("Problem 4: Social Media - Duplicate Account Detection")
print("-"*60)

def detect_duplicates(usernames):
    """Detect duplicate user accounts"""
    unique_users = set(usernames)
    duplicates = []

    for user in unique_users:
        count = usernames.count(user)
        if count > 1:
            duplicates.append(user)

    return len(unique_users), duplicates

# Test
users = ["alice", "bob", "charlie", "alice", "david", "bob"]
unique_count, dups = detect_duplicates(users)
print(f"Unique Accounts: {unique_count}")
print(f"Duplicate Accounts: {dups}")
print()
```

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(prices):
    premium = [price for price in prices if price > 2000]
    return premium

# Example
crop_prices = [1500, 2500, 1800, 3200]
premium_crops = filter_premium_crops(crop_prices)

print("Premium Crops:", premium_crops)
```

```python
# Problem 7: Application Health Checker

def check_system_health(errors):
    if errors == 0:
        return "Healthy"
    elif errors <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"

# Example
error_count = 7
status = check_system_health(error_count)

print("Error Count:", error_count)
print("System Status:", status)
```

```python
# Problem 8: Daily Transaction Limit Checker

def check_transaction(amount):
    daily_limit = 50000
    if amount <= daily_limit:
        return "Approved"
    else:
        return "Rejected"

# Example
transaction_amount = 60000
status = check_transaction(transaction_amount)

print("Transaction Amount:", transaction_amount)
print("Transaction Status:", status)
```

```python
# Problem 9: Student Attendance Eligibility

def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = sum(attendance_list)
    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return percentage, status

# Example (1 = Present, 0 = Absent)
attendance = [1,1,1,1,0,1,1,1,0,1]
percentage, eligibility = check_exam_eligibility(attendance)

print("Attendance Percentage:", percentage)
print("Exam Eligibility:", eligibility)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

# Example
units_consumed = 250
bill_amount, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", bill_amount)
print("Usage Level:", usage_status)
```
