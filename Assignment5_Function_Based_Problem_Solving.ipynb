```python
# Problem 3: Agriculture - Rainfall Adequacy Checker
print("Problem 3: Agriculture - Rainfall Adequacy Checker")
print("-"*60)

def check_rainfall(rainfall_data, required_rainfall=100):
    """Check if rainfall is adequate for crops"""
    total_rainfall = sum(rainfall_data)
    avg_rainfall = total_rainfall / len(rainfall_data)

    if total_rainfall >= required_rainfall:
        status = "Adequate Rainfall"
    else:
        status = "Insufficient Rainfall"

    return total_rainfall, avg_rainfall, status

# Test
rainfall = [25, 30, 28, 35]
total, avg, status = check_rainfall(rainfall)
print(f"Total Rainfall: {total} mm")
print(f"Average Rainfall: {avg:.2f} mm")
print(f"Status: {status}")
print()

# Problem 4: Social Media - Duplicate Account Detection
print("Problem 4: Social Media - Duplicate Account Detection")
print("-"*60)

def detect_duplicates(usernames):
    """Detect duplicate user accounts"""
    unique_users = set(usernames)
    duplicates = []

    for user in unique_users:
        count = usernames.count(user)
        if count > 1:
            duplicates.append(user)

    return len(unique_users), duplicates

# Test
users = ["alice", "bob", "charlie", "alice", "david", "bob"]
unique_count, dups = detect_duplicates(users)
print(f"Unique Accounts: {unique_count}")
print(f"Duplicate Accounts: {dups}")
print()
```
